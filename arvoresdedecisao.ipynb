```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = dataframe.copy()
```

```python
df
```

```python
df['cliente'] = df['cliente'].str.replace(r'\D','')
df['pedido'] = df['pedido'].str.replace(r'\D','')
df['produto'] = df['produto'].str.replace(r'\D','')
df['vendedor'] = df['vendedor'].str.replace(r'\D','')
df['review'] = df['review'].str.replace(r'\D','')
df
```

```python
result_df = pd.get_dummies(df,columns = ["tipopagamento"])
#getdummies separa todas as variações de uma só coluna e as separa em novas colunas, com uma contagem específica da vez que ela aparece relacionada a um cliente na linha do seu ID
```

```python
result_df
```

```python
reserva = result_df.copy()
reserva
```

```python
reserva = reserva.drop(reserva[reserva.score >3].index)
reserva

#ver quantidade de boas avaliações
```

```python
#dados de treino e teste:
# 70% para treino e 30% para teste
```

```python
y = result_df['score']
```

```python
X = result_df.drop('score', axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.3)
```

```python
#número de amostras para cada processo

X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
#número de targets para cada processo
```

```python
cls = DecisionTreeClassifier()
#classificador
```

```python
cls = cls.fit(X_train, y_train)
```

```python
cls.feature_importances_
```

```python
for feature, importancia in zip(result_df.columns, cls.feature_importances_):
  print("{}:{:.1f}%".format(feature, ((importancia*100))))
```

```python
result = cls.predict(X_test)
result
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, result))

#result guarda o que o algoritmo "acha que seria", enquanto o resultado está guardado em outro lugar. nesse print, ele mostra  que o algoritmo achou com o que de fato é
```

```python
from sklearn.model_selection import cross_val_score
allScores = cross_val_score(cls,X_train, y_train, cv=10)
allScores.mean()

# cross_val_score é uma ferramenta útil para avaliar e comparar diferentes modelos de aprendizado de máquina, permitindo que os cientistas de dados ajustem seus modelos para obter o melhor desempenho possível.
```
