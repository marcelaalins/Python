```python
notas.Nota.median()


#diferente da média (que pondera um valor central com base na quantidade de dados), a mediana verifica qual o valor central de um conjunto de dados.
#exemplo: C ={1,2,3,4}
#resultado:
#média: (1+2+3+4)/4 = 2.5
#mediana: como há dois números no centro do conjunto de dados, fazemos a soma de ambos e dividimos por 2. Então: 2+3 = 5 ---- 5/2 = 2.5
```

```python
print("Média:", notas["Nota"].mean())
print("Mediana:", notas["Nota"].median())
```

```python
tmdb = pd.read_csv("tmdb_5000_movies.csv")
tmdb
```

```python
tmdb.original_language.unique()
#variável categórica nominal
```

```python
tmdb.original_language.value_counts()
```

```python
tmdb.original_language.value_counts().index
#índice
```

```python
tmdb.original_language.value_counts().values
#valores atribuídos ao índice com base na contagem
```

```python
tmdb.original_language.value_counts().to_frame()

#usando o to_frame para criar uma tabela (dataframe)
```

```python
contagem_de_linguas = tmdb.original_language.value_counts().to_frame().reset_index()
#reset index é utilizado para transformar o índice (que é a referência das informações contidas na linha da tabela) numa coluna, trocando assim a sua referência para um índice automático (ID da linha)
```

```python
contagem_de_linguas.columns = ["original_language", "Total"]
contagem_de_linguas
```

```python
sns.barplot(x = "original_language", y= "Total", data = contagem_de_linguas)
```

```python
sns.catplot(x ="original_language", kind = "count", data = tmdb)


#função disponível apenas à partir da V. 0.9.0 do Seaborn
#forma mais simplificada de fazer o plot do gráfico sem precisar separar, contar e agrupar valores em várias etapas
```

```python
plt.pie(contagem_de_linguas["Total"], labels = contagem_de_linguas["original_language"])
```

```python
total_por_lingua = tmdb["original_language"].value_counts()
total_geral = total_por_lingua.sum()
total_de_ingles = total_por_lingua.loc["en"]
total_do_resto = total_geral - total_de_ingles
print(total_de_ingles, total_do_resto)
```

```python
dados = {
    'língua' : ["inglês", "outros"],
    'total' :  [total_de_ingles, total_do_resto]
}

dados

#separei os dados do nosso dataframe em categorias e renomeei as categorias
#como a proporção da língua inglesa é muito superior às das outras linguagens, é mais interessante referenciar esse baixo volume de outras línguas como "outros"
#em termos visuais, essa configuração é mais relevante
```

```python
dados = pd.DataFrame(dados)
dados
```

```python
sns.barplot(x= "língua", y = "total", data = dados)
```

```python
plt.pie(dados["total"], labels = dados["língua"])
```

```python
tmdb.query("original_language == 'en'")

#procurando, dentro da coluna original language, quais as ocorrências na linguagem inglesa (en)
```

```python
tmdb.query("original_language != 'en'")
#procurando informações que não tenham como linguagem original o inglês
```

```python
total_por_lingua_de_outros_filmes = tmdb.query("original_language != 'en'").original_language.value_counts()

#total de filmes com outras línguas que não sejam inglês
```

```python
filmes_sem_lingua_original_em_ingles = tmdb.query("original_language != 'en'").original_language.value_counts()
```

```python
filmes_sem_lingua_original_em_ingles = tmdb.query("original_language != 'en'")
sns.catplot(x = "original_language", kind = "count", data = filmes_sem_lingua_original_em_ingles)
```

```python
filmes_sem_lingua_original_em_ingles = tmdb.query("original_language != 'en'")
plt.figure(figsize=(5,10))
sns.catplot(x = "original_language", kind = "count", data = filmes_sem_lingua_original_em_ingles)
```

```python
#usando o aspect para plotar gráficos com espaços entre os nomes da coluna, de forma que fique legível

filmes_sem_lingua_original_em_ingles = tmdb.query("original_language != 'en'")
sns.catplot(x = "original_language", kind = "count", data = filmes_sem_lingua_original_em_ingles,
            aspect=2)
```

```python
#ordenando os dados do maior para o menor no gráfico usando order

#no order, eu carreguei uma tabela temporária que continha a contagem em ordem decrescente da quantidade de filmes por língua
#nessa tabela temporária, eu usei o index (que é o índice que identifica a linha, no caso, este é o tipo de linguagem do filme), e pedi para ordenar com base nisto. como já estava ordenado de forma decrescente, não precisei fazer nenhum outro tipo de tratamento
filmes_sem_lingua_original_em_ingles = tmdb.query("original_language != 'en'")
sns.catplot(x = "original_language", kind = "count", data = filmes_sem_lingua_original_em_ingles,
            aspect=2,
            order = total_por_lingua_de_outros_filmes.index)
```

```python
#usando uma paleta de cores no gráfico
filmes_sem_lingua_original_em_ingles = tmdb.query("original_language != 'en'")
sns.catplot(x = "original_language", kind = "count", data = filmes_sem_lingua_original_em_ingles,
            aspect=2,
            palette = "GnBu_d",
            order = total_por_lingua_de_outros_filmes.index)
```
